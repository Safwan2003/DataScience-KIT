## cleaning theMovies colum

```python
df['MOVIES'] = df['MOVIES'].str.strip()
```

### cleaning the votes

```python
df.isnull().sum()
```

```python
df['VOTES']=df['VOTES'].fillna('0')
```

```python
df['VOTES']=df['VOTES'].str.replace(',','')
```

```python
df['VOTES']=df['VOTES'].astype(int)
```

```python
df.info()
```

Rating cleaning

```python
df['RATING']=df['RATING'].fillna(0.0)
```

```python
df.isnull().sum()
```

Genre column cleaning

```python
df['GENRE']=df['GENRE'].str.strip('\n')
```

```python
df[['Type1','Type2','Type3']]=df['GENRE'].str.split(',',2,expand=True)
```
